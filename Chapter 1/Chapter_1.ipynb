## BA1E

```
text = input()
k = int(input())
L = int(input())
t = int(input())

def ClumpFindingProblem(dna, kk, LL, tt):
    solution = set()
    for i in range(len(dna)-L+1):
        ans={}
        single_window = dna[i:LL+i]
        # print(single_window)
        for j in range(len(single_window)-k+1):
            single_pattern = single_window[j:k+j]
            if single_pattern in ans.keys():
                ans[single_pattern]+=1
            else:
                ans[single_pattern]=1
        for key, value in ans.items():
            if value>=tt:
                solution.add(key)
    return solution

patterList = ClumpFindingProblem(text,k,L,t)
sol = ''
for item in patterList:
    sol+= str(item)+' '
print(sol)
```

## BA1K

```
def NumberToSymbol(Number):
    if Number == 0:
        return 'A'
    elif Number == 1:
        return 'C'
    elif Number == 2:
        return 'G'
    elif Number == 3:
        return 'T'
    
def NumberToPattern(n,k):
    if k==1:
        return NumberToSymbol(n)

    return NumberToPattern(n//4, k-1) + NumberToSymbol(n%4)
```

```
text = input()
k = int(input())

def ComputingFrequencyArray(Dna,k):
    ans = ''
    for i in range(4**k):
        pattern_i = NumberToPattern(i,k)
        
        d = PatternCount(Dna,pattern_i)
        ans+=str(d)+' '
        # print(pattern_i,d)
    return ans


sol = ComputingFrequencyArray(text,k)

print(sol)
```

## BA1L

```
pattern = input()

def SymbolToNumber(Word):
    if Word == 'A':
        return 0
    elif Word == 'C':
        return 1
    elif Word == 'G':
        return 2
    elif Word == 'T':
        return 3
    
def PatternToNumber(Pattern):
    if len(Pattern)==0:
        return 0
    symbol = Pattern[-1:]
    Prefix = Pattern[0:-1]
    return 4*PatternToNumber(Prefix) + SymbolToNumber(symbol)

result = PatternToNumber(pattern)

print(result)
```
