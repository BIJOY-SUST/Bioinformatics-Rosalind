```
import random
```

```
def Score(Motifs):
    sum_of_score = 0
    for i in range(len(Motifs[0])):
        sin_column = [motif2[i] for motif2 in Motifs]
        a = sin_column.count('A')
        c = sin_column.count('C')
        g = sin_column.count('G')
        t = sin_column.count('T')
        sum_of_score += (len(sin_column)-max(a,c,g,t))
    return sum_of_score
```

```
def ProfileMostProbableKMer(text,k,arr):
    result=0.0
    result_string = {}
    for i in range(len(text)-k+1):
        sin_string = text[i:i+k]
        sin_result=1.0
        for j in range(len(sin_string)):
            ch = sin_string[j]
            if ch == 'A':
                sin_result=sin_result*arr[0][j]
            elif ch == 'C':
                sin_result=sin_result*arr[1][j]
            elif ch == 'G':
                sin_result=sin_result*arr[2][j]
            elif ch == 'T':
                sin_result=sin_result*arr[3][j]
        # print(sin_string, sin_result)
        if sin_string in result_string.keys():
            if sin_result>=result_string[sin_string]:
                result_string[sin_string] = sin_result
        else:
            result_string[sin_string] = sin_result
    result_string = sorted(result_string.items(), key=lambda item: item[1], reverse=True)
    result_string = result_string[0]
    return result_string[0]
```

```
def RandomizedMotifSearch(Dna, k, t):
    
    Motifs = []
    for i in range(t):
        n = len(Dna[i])
        k1 = random.randint(0, n-k)
        motif_i = Dna[i][k1:k1+k]
        Motifs.append(motif_i)

    bestmotifs = Motifs
    # print(bestmotifs)
    cntt = 0
    while 1==1:

        profile = []
        nucleotide = ['A','C','G','T']
        for ch in nucleotide:
            sin_profile = []
            for ii in range(k):
                s_mat = []
                for s_motif in Motifs:
                    # print(ch,ii,s_motif)
                    s_mat.append(s_motif[ii])
                result = (s_mat.count(ch)+1)/(len(Motifs)+4)
                sin_profile.append(result)
            profile.append(sin_profile)

        for i in range(t):
            Motifs[i] = ProfileMostProbableKMer(Dna[i], k, profile)

        if Score(Motifs) < Score(bestmotifs):
            bestmotifs = Motifs
        else:
            return bestmotifs
        
```

```
k = int(input("k-mers length: "))
t = int(input("Enter t: "))
Dna=[]
for i in range(t):
    text = input()
    Dna.append(text)


bestmotifs = RandomizedMotifSearch(Dna, k, t)
finalscore = Score(bestmotifs)

for i in range(int(1000)):
    motifs = RandomizedMotifSearch(Dna, k, t)
    score = Score(motifs)
    if finalscore > score:
        finalscore = score
        bestmotifs = motifs

print('A collection BestMotifs resulting from running RandomizedMotifSearch(Dna, k, t) 1000 times:')
for item in bestmotifs:
    print(item)
```

```
def ProfileRandomlyProbableKMer(text,k,arr):
    
    result_string = {}
    for i in range(len(text)-k+1):
        sin_string = text[i:i+k]
        sin_result=1.0
        for j in range(len(sin_string)):
            ch = sin_string[j]
            if ch == 'A':
                sin_result=sin_result*arr[0][j]
            elif ch == 'C':
                sin_result=sin_result*arr[1][j]
            elif ch == 'G':
                sin_result=sin_result*arr[2][j]
            elif ch == 'T':
                sin_result=sin_result*arr[3][j]
        # print(sin_string, sin_result)
        result_string[sin_string] = sin_result
    sum = 0.0
    val_list = []
    val_aclu_list = []
    for key,val in result_string.items():
        sum+=val
        val_list.append(val)
        val_aclu_list.append(sum)
    
    prpkm = random.random()
    indx = -1
    for key,val in result_string.items():
        indx+=1
        if val_aclu_list[indx]/sum > prpkm:
            return key
```

```
def GibbsSampler(Dna, k, t, N):
    
    Motifs = []
    for i in range(t):
        n = len(Dna[i])
        k1 = random.randint(0, n-k)
        motif_i = Dna[i][k1:k1+k]
        Motifs.append(motif_i)

    bestmotifs = Motifs
    # print(bestmotifs)

    for cntt in range(N):
        k1 = random.randint(0, t-1)

        newmotifs = []
        for jj in range(len(Motifs)):
            if jj!=k1:
                newmotifs.append(Motifs[jj])
        
        profile = []
        nucleotide = ['A','C','G','T']
        for ch in nucleotide:
            sin_profile = []
            for ii in range(k):
                s_mat = []
                for s_motif in newmotifs:
                    # print(ch,ii,s_motif)
                    s_mat.append(s_motif[ii])
                result = (s_mat.count(ch)+1)/(len(newmotifs)+4)
                sin_profile.append(result)
            profile.append(sin_profile)
        
        Motifs[k1] = ProfileRandomlyProbableKMer(Dna[k1], k, profile)
        

        if Score(Motifs) < Score(bestmotifs):
            bestmotifs = Motifs
        else:
            return bestmotifs
        
```

```
k = int(input("k-mers length: "))
t = int(input("Enter t: "))
N = int(input("Enter N: "))
Dna=[]
for i in range(t):
    text = input()
    Dna.append(text)


bestmotifs = GibbsSampler(Dna, k, t, N)
finalscore = Score(bestmotifs)

for i in range(int(20)):
    motifs = GibbsSampler(Dna, k, t, N)
    score = Score(motifs)
    if finalscore > score:
        finalscore = score
        bestmotifs = motifs

print('The strings BestMotifs resulting from running GibbsSampler(Dna, k, t, N) with 20 random starts:')
for item in bestmotifs:
    print(item)
```

```
def Compositionk(k,text):
    result = []
    for i in range(len(text)-k+1):
        read = text[i:i+k]
        # if read in result:
        #     continue
        result.append(read)
    return result
```

```
k = int(input("k-mers length: "))
text = input("Enter the string: ")


Results = Compositionk(k,text)

Results = sorted(Results, key=lambda x: x, reverse=False)

print('Compositionk(Text) (the k-mers can be provided in any order):')
for item in Results:
    print(item)
```

```
def StringSpelled2(Motifs):
    result = ''
    for i in range(len(Motifs)):
        for j in range(len(Motifs)):
            motif1 = Motifs[i]
            motif2 = Motifs[j]
            if i!=j and motif1[1:]==motif2[:-1]:
                result+=motif1+' -> '+motif2+'\n'
    return result
```

```
motifs=[]
for i in range(5):
    text = input()
    motifs.append(text)

Results = StringSpelled2(motifs)

print('The overlap graph Overlap(Patterns) :')
print(Results)
```

## Task 6

```
def DeBruijn(dna1,k):
    # print(dna1,k)
    result = {}
    for i in range(len(dna1)-k+1):
        a = dna1[i:i+k-1]
        b = dna1[i+1:i+k]
        # print(a,b,i)
        if a in result.keys():
            result[a] += ','+b
        else:
            result[a] = b

    return result
```

```
k = int(input("k-mers length: "))

dna1 = input("DNA string: ")

Results = DeBruijn(dna1,k)

print('DeBruijnk(Text), in the form of an adjacency list:')
for key in sorted(Results):
    print(key,' -> ',Results[key])
```

```
def DeBruijnLarge(Motifs):
    # print(dna1,k)
    result = {}
    for dna1 in Motifs:
        a = dna1[:-1]
        b = dna1[1:]
        # print(a,b,i)
        if a in result.keys():
            result[a] += ','+b
        else:
            result[a] = b

    return result
```

```
motifs=[]
for i in range(7):
    text = input()
    motifs.append(text)

Results = DeBruijnLarge(motifs)

print('The de Bruijn graph DeBruijn(Patterns), in the form of an adjacency list:')
for key in sorted(Results):
    print(key,' -> ',Results[key])
```
