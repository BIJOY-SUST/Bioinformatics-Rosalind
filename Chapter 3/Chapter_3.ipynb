```
def StringSpelled(Motifs):
    result = ''
    for i in range(len(Motifs)):
        for j in range(len(Motifs)):
            motif1 = Motifs[i]
            motif2 = Motifs[j]
            if i!=j and motif1[1:]==motif2[:-1]:
                result+=motif1+' -> '+motif2+'\n'
    return result
```

```
motifs=[]
for i in range(5):
    text = input()
    motifs.append(text)

Results = StringSpelled(motifs)

print(Results)
```

## BA3D

```
def DeBruijn(dna1,k):
    # print(dna1,k)
    result = {}
    for i in range(len(dna1)-k+1):
        a = dna1[i:i+k-1]
        b = dna1[i+1:i+k]
        # print(a,b,i)
        if a in result.keys():
            result[a] += ','+b
        else:
            result[a] = b

    return result
```

```
k = int(input())

dna1 = input()

Results = DeBruijn(dna1,k)

for key in sorted(Results):
    print(key,' -> ',Results[key])
```

```
def DeBruijnLarge(Motifs):
    # print(dna1,k)
    result = {}
    for dna1 in Motifs:
        a = dna1[:-1]
        b = dna1[1:]
        # print(a,b,i)
        if a in result.keys():
            result[a] += ','+b
        else:
            result[a] = b

    return result
```

```
def HierholzerAlgorithm(adj):

    edge_count = dict() 
    for i in range(len(adj)):
        edge_count[i] = len(adj[i])

    empt = []
    if len(adj) == 0:
        return empt

    curr_path = [] 
    circuit = [] 
    curr_path.append(0) 
    curr_v = 0 
    while len(curr_path):
        if edge_count[curr_v]: 
            curr_path.append(curr_v) 
            next_v = adj[curr_v][-1] 
            edge_count[curr_v] -= 1
            adj[curr_v].pop() 
            curr_v = next_v 
        else: 
            circuit.append(curr_v) 
            curr_v = curr_path[-1] 
            curr_path.pop() 
    return circuit
```

```
def get_balance_count(adj_list):
    balanced_count = dict.fromkeys(adj_list.keys(), 0)
    for node in adj_list.keys():
        for out in adj_list[node]:
            balanced_count[node] -= 1
            try:
                balanced_count[out] += 1
            except:
                balanced_count[out] = 1
    return balanced_count
```

```
def eulerian_path_problem(dict):
    stack=[]
    balanced_count = get_balance_count(dict)
    stack.append([k for k, v in balanced_count.items() if v==-1][0])
    path = []
    while stack != []:
        u_v = stack[-1]
        try:
            w = dict[u_v][0]
            stack.append(w)
            dict[u_v].remove(w)
        except:
            path.append(stack.pop())
    return path[::-1]
```

```
def genome_path_problem(kmers, apppend_last=True):
    genome = ''
    kmer_length = len(kmers[0])
    for kmer in kmers:
        genome += kmer[0]
    if apppend_last:
        genome += kmer[1:]
    return genome
```

```
def suffix_composition(k, text, uniq=False):
    kmers = []
    for i in range(len(text)+1-k):
        kmers.append(text[i:i+k-1])
    if uniq:
        return sorted(list(kmers))
    else:
        return sorted(kmers)
```

```
def debrujin_graph_from_kmers(patterns):
    kmers = []
    for pattern in patterns:
        kmers = kmers + suffix_composition(len(pattern), pattern, uniq=True)
    kmers = set(kmers)
    dict = {}
    for kmer1 in kmers:
        dict[kmer1] = []
    for kmer in patterns:
        dict[kmer[0:-1]].append(kmer[1:])
    return dict
```

```
def StringReconstructionProblem(k,patterns):
    result = debrujin_graph_from_kmers(patterns)
    result = eulerian_path_problem(result)
    result = genome_path_problem(result)

    return result
```

```
k = int(input())
data = []
for i in range(6):
    dd = input()
    data.append(dd)
result = StringReconstructionProblem(k, data)
print(result)
```

```
import numpy as np

def Reconstructstring(k, d, mers):
    mers = np.vstack(mers)

    ppp = mers[0, 0] + ''.join([s[-1:] for s in mers[1:, 0]])
    sss = mers[0, 1] + ''.join([s[-1:] for s in mers[1:, 1]])
    
    result = ppp[0: d+k] + sss
    
    return result
```

```
k = int(input())
d = int(input())
ss = []
for i in range(9):
    hh = input()
    ss.append(hh)

mers = [s.split('|') for s in ss]
ans = Reconstructstring(k, d, mers)
ans
```
