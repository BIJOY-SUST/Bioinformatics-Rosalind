## BA2A

```
def HammingDistance(p, q):
    ans=0
    for i in range(len(p)):
        if p[i]!=q[i]:
            ans+=1
    return ans
```

```
def NumberToSymbol(Number):
    if Number == 0:
        return 'A'
    elif Number == 1:
        return 'C'
    elif Number == 2:
        return 'G'
    elif Number == 3:
        return 'T'
    
def NumberToPattern(n,k):
    if k==1:
        return NumberToSymbol(n)

    return NumberToPattern(n//4, k-1) + NumberToSymbol(n%4)
```

```
def SymbolToNumber(Word):
    if Word == 'A':
        return 0
    elif Word == 'C':
        return 1
    elif Word == 'G':
        return 2
    elif Word == 'T':
        return 3
    
def PatternToNumber(Pattern):
    if len(Pattern)==0:
        return 0
    symbol = Pattern[-1:]
    Prefix = Pattern[0:-1]
    return 4*PatternToNumber(Prefix) + SymbolToNumber(symbol)
```

```
def Neighbors(Dna,d):
    ans = []
    for i in range(4**len(Dna)):
        pattern_i = NumberToPattern(i,len(Dna))
        
        hh = HammingDistance(Dna,pattern_i)
        if hh<=d:
            ans.append(pattern_i)
        # print(pattern_i,d)
    return ans
```

```
def MotifEnumeration(dna, k, d):
    answers = []
    for item in dna:
        pattern = set()
        for i in range(len(item)-k+1):
            kmers = Neighbors(item[i:i+k], d)
            kmers = set(kmers)
            for kmer in kmers:
                pattern.add(kmer)
        for kmer in pattern:
            answers.append(kmer)
    motifPattern = []
    for item in answers:
        if answers.count(item) == len(dna):
            motifPattern.append(item)
    motifPattern = list(set(motifPattern))
    return motifPattern
```

```
k = int(input())
d = int(input())
dna = []
for i in range(4):
    dd = input()
    dna.append(dd)

Results = MotifEnumeration(dna, k, d)
sol = ''
for item in Results:
    sol+= str(item)+' '
print(sol)
```

## BA2B

```
def d(pattern, Dna):
    d_final = 0
    for sin_dna in Dna:
        h_dis = len(pattern)+1
        for j in range(len(sin_dna)-len(pattern)+1):
            single_pattern = sin_dna[j:len(pattern)+j]
            d_hamming = HammingDistance(pattern, single_pattern)
            if h_dis>d_hamming:
                h_dis = d_hamming
            
        d_final+=h_dis
    
    return d_final
```

```
k = int(input())

dna1 = input()
dna2 = input()
dna3 = input()
dna4 = input()
dna5 = input()

Dna = [dna1,dna2,dna3,dna4,dna5]


def MedianString(Dna,k):
    distance = 1e18
    ans = ''
    for i in range(4**k):
        pattern_i = NumberToPattern(i,k)
        sin_d = d(pattern_i,Dna)
        if distance>sin_d:
            distance = sin_d
            ans = pattern_i
    return ans


answer = MedianString(Dna,k)

print(answer)
```

```
def ProfileMostProbableKMer(text,k,arr):
    result=0.0
    result_string = {}
    for i in range(len(text)-k+1):
        sin_string = text[i:i+k]
        sin_result=1.0
        for j in range(len(sin_string)):
            ch = sin_string[j]
            if ch == 'A':
                sin_result=sin_result*arr[0][j]
            elif ch == 'C':
                sin_result=sin_result*arr[1][j]
            elif ch == 'G':
                sin_result=sin_result*arr[2][j]
            elif ch == 'T':
                sin_result=sin_result*arr[3][j]
        # print(sin_string, sin_result)
        result_string[sin_string] = sin_result
    result_string = sorted(result_string.items(), key=lambda item: item[1], reverse=True)
    result_string = result_string[0]
    return result_string[0]
```

```
text = input()
k = int(input())
rows, cols = (4, k) 
arr = [[0 for i in range(cols)] for j in range(rows)] 
for i in range(4):
    dd = input()
    dd = dd.split(' ')
    for j in range(len(dd)):
        arr[i][j] = float(dd[j])

Results = ProfileMostProbableKMer(text, k, arr)

print(Results)
```

```
def Score(Motifs):
    sum_of_score = 0
    for i in range(len(Motifs[0])):
        sin_column = [motif2[i] for motif2 in Motifs]
        a = sin_column.count('A')
        c = sin_column.count('C')
        g = sin_column.count('G')
        t = sin_column.count('T')
        sum_of_score += (len(sin_column)-max(a,c,g,t))
    return sum_of_score
```

```
def GreedyMotifSearch(Dna, k, t):
    bestmotifs = [item[:k] for item in Dna]
    firststring = Dna[0]
    for i in range(len(firststring)-k+1):
        motif1 = firststring[i:i+k]
        Motifs = [motif1]
        # print(motif1)

        for j in range(1,t):
            motif = Motifs[:j]
            # print(motif)
            profile = []
            nucleotide = ['A','C','G','T']
            for ch in nucleotide:
                sin_profile = []
                for ii in range(len(motif1)):
                    s_mat = []
                    for s_motif in motif:
                        # print(ch,ii,s_motif)
                        s_mat.append(s_motif[ii])
                    result = s_mat.count(ch)/len(motif)
                    sin_profile.append(result)
                profile.append(sin_profile)
            # print(Dna[j], k, profile[0][0])
            # print(Dna[j], k, profile)

            p_m_p_kmer = ProfileMostProbableKMer(Dna[j], k, profile)
            # print(p_m_p_kmer)
            Motifs.append(p_m_p_kmer)
        # print(bestmotifs)
        # print(Score(bestmotifs))
        # print(Motifs)
        # print(Score(Motifs))

        if Score(Motifs) < Score(bestmotifs):
            bestmotifs = Motifs
            
    return bestmotifs
        
```

```
def GibbsSampler(Dna, k, t, N):
    
    Motifs = []
    for i in range(t):
        n = len(Dna[i])
        k1 = random.randint(0, n-k)
        motif_i = Dna[i][k1:k1+k]
        Motifs.append(motif_i)

    bestmotifs = Motifs
    # print(bestmotifs)

    for cntt in range(N):
        k1 = random.randint(0, t-1)

        newmotifs = []
        for jj in range(len(Motifs)):
            if jj!=k1:
                newmotifs.append(Motifs[jj])
        
        profile = []
        nucleotide = ['A','C','G','T']
        for ch in nucleotide:
            sin_profile = []
            for ii in range(k):
                s_mat = []
                for s_motif in newmotifs:
                    # print(ch,ii,s_motif)
                    s_mat.append(s_motif[ii])
                result = (s_mat.count(ch)+1)/(len(newmotifs)*2)
                sin_profile.append(result)
            profile.append(sin_profile)
        
        Motifs[k1] = ProfileRandomlyProbableKMer(Dna[k1], k, profile)
        

        if Score(Motifs) < Score(bestmotifs):
            bestmotifs = Motifs
        else:
            return bestmotifs
        
```

```
k = int(input())
t = int(input())
N = int(input())
Dna=[]
for i in range(t):
    text = input()
    Dna.append(text)


bestmotifs = GibbsSampler(Dna, k, t, N)
finalscore = Score(bestmotifs)

for i in range(int(20)):
    motifs = GibbsSampler(Dna, k, t, N)
    score = Score(motifs)
    if finalscore > score:
        finalscore = score
        bestmotifs = motifs

for item in bestmotifs:
    print(item)
```

## BA2H

```
def DistanceBetweenPatternAndStrings(pattern, dna):
    k = len(pattern)
    distance = 0
    for each_string in dna:
        haming_distance = k+1
        for i in range(len(each_string)-k+1):
            dd = HammingDistance(each_string[i:i+k], pattern)
            if haming_distance > dd:
                haming_distance = dd
        distance = distance + haming_distance
    return distance
```

```
pattern = input()
dna = input()
dna = dna.split(' ')

result = DistanceBetweenPatternAndStrings(pattern, dna)
print(result)
```
