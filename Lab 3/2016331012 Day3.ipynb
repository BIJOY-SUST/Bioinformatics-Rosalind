```
pattern = input("A DNA string Genome: ")

def MinimumSkewProblem(text):
    ans = []
    c=0
    g=0
    min_skew = 0
    for i in range(len(text)):
        if text[i]=='C':
            c+=1
        elif text[i]=='G':
            g+=1
        if g-c < min_skew:
            ans=[]
            ans.append(i+1)
            min_skew = g-c
        elif g-c==min_skew:
            ans.append(i+1)


    return ans
reverse_complement = MinimumSkewProblem(pattern)
print('Following position in a genome minimizing the skew: ',reverse_complement)
```

## Task 2

```
text = input("A DNA full string: ")
k = int(input("k-mers length: "))
L = int(input("Window Length: "))
t = int(input("Pattern appears at least: "))

def ClumpFindingProblem(dna, kk,LL,tt):
    solution = set()
    for i in range(len(dna)-L+1):
        ans={}
        single_window = dna[i:i+LL]
        # print(single_window)
        for j in range(len(single_window)-k+1):
            single_pattern = single_window[j:k+j]
            if single_pattern in ans.keys():
                ans[single_pattern]+=1
            else:
                ans[single_pattern]=1
        for key, value in ans.items():
            if value>=tt:
                solution.add(key)
    return solution

patterList = ClumpFindingProblem(text,k,L,t)

print('Following patterns forming clumps in a string: ', patterList)
```

## Task 3

##### Distance Calculation

```
def HammingDistance(p, q):
    ans=0
    for i in range(len(p)):
        if p[i]!=q[i]:
            ans+=1
    return ans
```

```
def d(pattern, Dna):
    d_final = 0
    for sin_dna in Dna:
        h_dis = len(pattern)+1
        for j in range(len(sin_dna)-len(pattern)+1):
            single_pattern = sin_dna[j:len(pattern)+j]
            d_hamming = HammingDistance(pattern, single_pattern)
            if h_dis>d_hamming:
                h_dis = d_hamming
            
        d_final+=h_dis
    
    return d_final
```

##### NumberToPattern

```
def NumberToSymbol(Number):
    if Number == 0:
        return 'A'
    elif Number == 1:
        return 'C'
    elif Number == 2:
        return 'G'
    elif Number == 3:
        return 'T'
    
def NumberToPattern(n,k):
    if k==1:
        return NumberToSymbol(n)

    return NumberToPattern(n//4, k-1) + NumberToSymbol(n%4)
```

##### Main Code

```
k = int(input("k-mers length: "))

dna1 = input("1st DNA string: ")
dna2 = input("2nd DNA string: ")
dna3 = input("3rd DNA string: ")
dna4 = input("4th DNA string: ")
dna5 = input("5th DNA string: ")

Dna = [dna1,dna2,dna3,dna4,dna5]


def MedianString(Dna,k):
    distance = 1e18
    ans = ''
    for i in range(4**k):
        pattern_i = NumberToPattern(i,k)
        sin_d = d(pattern_i,Dna)
        if distance>sin_d:
            distance = sin_d
            ans = pattern_i
    return ans


answer = MedianString(Dna,k)

print('The desired Median String is: ',answer)
```

## Task 4

```
def ProfileMostProbableKMer(text,k,arr):
    result=0.0
    result_string = {}
    for i in range(len(text)-k+1):
        sin_string = text[i:i+k]
        sin_result=1.0
        for j in range(len(sin_string)):
            ch = sin_string[j]
            if ch == 'A':
                sin_result=sin_result*arr[0][j]
            elif ch == 'C':
                sin_result=sin_result*arr[1][j]
            elif ch == 'G':
                sin_result=sin_result*arr[2][j]
            elif ch == 'T':
                sin_result=sin_result*arr[3][j]
        # print(sin_string, sin_result)
        if sin_string in result_string.keys():
            if sin_result>=result_string[sin_string]:
                result_string[sin_string] = sin_result
        else:
            result_string[sin_string] = sin_result
            
    result_string = sorted(result_string.items(), key=lambda item: item[1], reverse=True)
    result_string = result_string[0]
    return result_string[0]
```

```
text = input("A DNA full string: ")
k = int(input("k-mers length: "))
rows, cols = (4, k) 
arr = [[0 for j in range(cols)] for i in range(rows)] 
for i in range(4):
    for j in range(k):
        val = float(input())
        arr[i][j] = val


Results = ProfileMostProbableKMer(text, k, arr)

print('A Profile-most probable k-mer in Text: ',Results)
```

## Task 5

```
def Score(Motifs):
    sum_of_score = 0
    for i in range(len(Motifs[0])):
        sin_column = [motif2[i] for motif2 in Motifs]
        a = sin_column.count('A')
        c = sin_column.count('C')
        g = sin_column.count('G')
        t = sin_column.count('T')
        sum_of_score += (len(sin_column)-max(a,c,g,t))
    return sum_of_score
```

```
def GreedyMotifSearch(Dna, k, t):
    bestmotifs = [item[:k] for item in Dna]
    firststring = Dna[0]
    for i in range(len(firststring)-k+1):
        motif1 = firststring[i:i+k]
        Motifs = [motif1]
        # print(motif1)

        for j in range(1,t):
            motif = Motifs[:j]
            # print(motif)
            profile = []
            nucleotide = ['A','C','G','T']
            for ch in nucleotide:
                sin_profile = []
                for ii in range(k):
                    s_mat = []
                    for s_motif in motif:
                        # print(ch,ii,s_motif)
                        s_mat.append(s_motif[ii])
                    result = s_mat.count(ch)/len(motif)
                    sin_profile.append(result)
                profile.append(sin_profile)
            # print(Dna[j], k, profile[0][0])
            # print(Dna[j], k, profile)

            p_m_p_kmer = ProfileMostProbableKMer(Dna[j], k, profile)
            # print(p_m_p_kmer)
            Motifs.append(p_m_p_kmer)
        # print(bestmotifs)
        # print(Score(bestmotifs))
        # print(Motifs)
        # print(Score(Motifs))

        if Score(Motifs) < Score(bestmotifs):
            bestmotifs = Motifs
            
    return bestmotifs
        
```

```
k = int(input("k-mers length: "))
t = int(input("Enter t: "))
Dna=[]
for i in range(t):
    text = input()
    Dna.append(text)

Results = GreedyMotifSearch(Dna, k, t)

print('A collection of strings BestMotifs resulting from running GreedyMotifSearch(Dna, k, t):')
for item in Results:
    print(item)
```
