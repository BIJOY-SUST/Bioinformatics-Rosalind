```
def translation(RNA):
    answer = ''
    for i in range(0,len(RNA)-(3+(len(RNA)%3)),3):
        if rna_codon_table[RNA[i:i+3]]=='STOP':
            break
        else:
            answer+=rna_codon_table[RNA[i:i+3]]
    return answer
```

```
def reverse_complement(pp):
    rvstr = ''
    for ch in pp:
        if ch == 'A':
            rvstr+='T'
        elif ch=='T':
            rvstr+='A'
        elif ch=='G':
            rvstr+='C'
        elif ch=='C':
            rvstr+='G'
    return rvstr[::-1]
```

```
def dna_to_rna(pat):
    return pat.replace('T','U')
```

```
def PeptideEncoding(dna, peptides):
    answer = ''
    d = len(peptides)*3
    for i in range(len(dna)-d+1):
        patt = dna[i:i+d]
        patt2 = reverse_complement(patt)
        p1 = dna_to_rna(patt)
        p2 = dna_to_rna(patt2)

        a1=''
        a2=''
        for i in range(0,len(p1)-2,3):
            kee = p1[i:i+3]
            a1+=rna_codon_table[kee]
        if a1==peptides:
            answer = answer+patt+"\n"

        for i in range(0,len(p2)-2,3):
            kee = p2[i:i+3]
            a2+=rna_codon_table[kee]
        if a2==peptides:
            answer = answer+reverse_complement(patt2)+"\n"
        
    return answer
```

```
dna = input('Enter DNA string: ')
peptides = input("Enter Peptides: ")
result = PeptideEncoding(dna, peptides)
print(result)
```

```
def Cyclospectrum(amino_acid):
    answer = []
    answer.append(0)
    
    for i in range(1,len(amino_acid)):
        for j in range(len(amino_acid)):
            if i+j>len(amino_acid):
                d = i+j - len(amino_acid)
                st = amino_acid[j:]+amino_acid[:d]
                d = 0
                for ch in st:
                    d+=newcleotide_to_mass[ch]
                answer.append(d)
            else:
                st = amino_acid[j:j+i]
                d=0
                for ch in st:
                    d+=newcleotide_to_mass[ch]
                answer.append(d)
    d = 0
    for ch in amino_acid:
        d+=newcleotide_to_mass[ch]
    answer.append(d)

    answer.sort()
    sol = ''
    for item in answer:
        sol+=str(item)+ ' '
    return sol
```

```
amino_acid = input('Enter Amino Acid string: ')
result = Cyclospectrum(amino_acid)
print(result)
```

```
def get_balance_count(adj_list):
    balanced_count = dict.fromkeys(adj_list.keys(), 0)
    for node in adj_list.keys():
        for out in adj_list[node]:
            balanced_count[node] -= 1
            try:
                balanced_count[out] += 1
            except:
                balanced_count[out] = 1
    return balanced_count
```

```
def eulerian_path_problem(dict):
    stack=[]
    balanced_count = get_balance_count(dict)
    stack.append([k for k, v in balanced_count.items() if v==-1][0])
    path = []
    while stack != []:
        u_v = stack[-1]
        try:
            w = dict[u_v][0]
            stack.append(w)
            dict[u_v].remove(w)
        except:
            path.append(stack.pop())
    return path[::-1]
```

```
def suffix_composition(k, text, uniq=False):
    kmers = []
    for i in range(len(text)+1-k):
        kmers.append(text[i:i+k-1])
    if uniq:
        return sorted(list(kmers))
    else:
        return sorted(kmers)
```
