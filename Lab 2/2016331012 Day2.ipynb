## Task 1

```
pattern = input("A DNA string Pattern: ")


def ReverseComplement(text):
    ans = ''
    for i in range(len(text)):
        if text[i]=='A':
            ans+='T'
        elif text[i]=='T':
            ans+='A'
        elif text[i]=='G':
            ans+='C'
        elif text[i]=='C':
            ans+='G'

    ans = ans[::-1]
    return ans
reverse_complement = ReverseComplement(pattern)
print('Pattern, the reverse complement of Pattern: ',reverse_complement)
```

## Task 2

```
pattern1 = input("First DNA string Pattern: ")
pattern2 = input("Second DNA string Pattern: ")


def HammingDistance(p, q):
    ans=0
    for i in range(len(p)):
        if p[i]!=q[i]:
            ans+=1
    return ans

distance = HammingDistance(pattern1, pattern2)

print('An integer value representing the Hamming distance: ', distance)
```

## Task 3

```
pattern = input("A DNA string Pattern: ")
text = input("A DNA full string: ")
d = int(input("Enter number of most mismatches: "))

def ApproximatePatternMatching(Pattern, Text, d):
    positions = []
    for i in range(len(Text)-len(Pattern)+1):
        if HammingDistance(Pattern, Text[i:i + len(Pattern)]) <= d:
            positions.append(i)
    return positions

allApproximateOccurrences = ApproximatePatternMatching(pattern,text,d)

print('All starting positions where Pattern appears as a substring of Text with at most d mismatches: ', allApproximateOccurrences)
```

## Task 4

```
text = input("A DNA full string: ")
k = int(input("Enter pattern length: "))
d = int(input("Enter number of most mismatches: "))


def NearestPattern(Pattern, d, NearestWords):
    Root = ['A', 'C', 'G', 'T']
    for pi in range(len(Pattern)):
        for ri in range(len(Root)):
            NewPattern = Pattern[:pi] + Root[ri] + Pattern[pi+1:]
            if d <= 1:
                NearestWords.add(NewPattern)
            else:
                NearestPattern(NewPattern, d-1, NearestWords)

def MostFrequentWords(text, k, d):
    AllResults = {}
    for i in range(len(text) - k + 1):
        NearestWords = set()
        NearestPattern(text[i:i + k], d, NearestWords)
        
        for words in NearestWords:
            if words in AllResults:
                AllResults[words] += 1
            else:
                AllResults[words] = 1
    
    Results = []
    for item in AllResults.keys():
        if AllResults[item] == max(AllResults.values()):
            Results.append(item)
    
    return Results

Results = MostFrequentWords(text, k, d)

print('All most frequent k-mers with up to d mismatches in Text: ',Results)
```

## Task 5

```
text = input("A DNA full string: ")
k = int(input("Enter pattern length: "))
d = int(input("Enter number of most mismatches: "))


def NearestPattern(Pattern, d, NearestWords):
    Root = ['A', 'C', 'G', 'T']
    for pi in range(len(Pattern)):
        for ri in range(len(Root)):
            NewPattern = Pattern[:pi] + Root[ri] + Pattern[pi+1:]
            if d <= 1:
                NearestWords.add(NewPattern)
            else:
                NearestPattern(NewPattern, d-1, NearestWords)

def MostFrequentWords(text, k, d):
    AllResults = {}
    for i in range(len(text) - k + 1):
        NearestWords = set()
        NearestWordsReverse = set()
        NearestPattern(text[i:i + k], d, NearestWords)
        NearestPattern(ReverseComplement(text[i:i + k]), d, NearestWordsReverse)
        for words in NearestWords:
            if words in AllResults:
                AllResults[words] += 1
            else:
                AllResults[words] = 1
    
        for words in NearestWordsReverse:
            if words in AllResults:
                AllResults[words] += 1
            else:
                AllResults[words] = 1
    
        
    Results = []
    for item in AllResults.keys():
        if AllResults[item] == max(AllResults.values()):
            Results.append(item)
    
    return Results

Results = MostFrequentWords(text, k, d)

print('All most frequent k-mers with up to d mismatches in Text: ',Results)
```

## Task 6

```
pattern = input("A DNA string Pattern: ")

def SymbolToNumber(Word):
    if Word == 'A':
        return 0
    elif Word == 'C':
        return 1
    elif Word == 'G':
        return 2
    elif Word == 'T':
        return 3
    
def PatternToNumber(Pattern):
    if len(Pattern)==0:
        return 0
    symbol = Pattern[-1:]
    Prefix = Pattern[0:-1]
    return 4*PatternToNumber(Prefix) + SymbolToNumber(symbol)

result = PatternToNumber(pattern)

print('PatternToNumber: ',result)
```

## Task 7

```
n = int(input("Enter the number: "))
k = int(input("Enter the length: "))

def NumberToSymbol(Number):
    if Number == 0:
        return 'A'
    elif Number == 1:
        return 'C'
    elif Number == 2:
        return 'G'
    elif Number == 3:
        return 'T'
    
def NumberToPattern(n,k):
    if k==1:
        return NumberToSymbol(n)

    return NumberToPattern(n//4, k-1) + NumberToSymbol(n%4)

result = NumberToPattern(n,k)

print('NumberToPattern: ',result)
```
