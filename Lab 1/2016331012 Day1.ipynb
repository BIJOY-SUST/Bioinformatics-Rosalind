## Task 1

```
text = input("What is the genomic sequence? ")
pattern = input("What is the pattern? ")


def PatternCount(text, pattern):
    cnt = 0
    for i in range(len(text)-len(pattern)+1):
        temp_text = text[i:i+len(pattern)]
        if temp_text == pattern:
            cnt+=1
    return cnt

number_of_occurrences = PatternCount(text, pattern)
print(number_of_occurrences)
```

## Task 4

```
print("Input first matrix \n")
A=[]

r=int(input("Enter number of row "))
c=int(input("Enter number of column "))

print("Data?\n")
for i in range(r): 
    row=[]
    for j in range(c): 
        row.append(int(input())) 
    A.append(row)     


print(A)

print("Input second matrix \n")
B=[]

r=int(input("Enter number of row "))
c=int(input("Enter number of column "))

print("Data?\n")
for i in range(r): 
    row=[]          
    for j in range(c): 
        row.append(int(input()))  
    B.append(row)                


result = [[sum(a * b for a, b in zip(A_row, B_col))  for B_col in zip(*B)] for A_row in A] 

print("Result of multiplication of A and B \n")
for row in result: 
   print(row)
```
