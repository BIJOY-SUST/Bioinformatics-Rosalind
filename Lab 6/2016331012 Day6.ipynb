## Task 2

```
spectrum = input()
spectrum = spectrum.split()
data = []
for ch in spectrum:
    dd = int(ch)
    data.append(dd)
convolution = [i-j for i in data for j in data if i-j > 0]
con_dict = {}
for item in convolution:
    if item not in con_dict.keys():
        con_dict[item] = 1
    else:
        con_dict[item] += 1
con_dict = dict(sorted(con_dict.items(), key=lambda item: item[1], reverse=True))
# print(con_dict)

sol = ''
for key, val in con_dict.items():
    # print(key, val)
    for i in range(val):
        sol+=str(key)+' '
print(sol)
```

## Task 5

```
text = input()

lengthOfString = len(text)
sDict = {}
sDictr = {}

for i in range(lengthOfString):
    pp = text[i:lengthOfString]
    sDict[i]=pp
    sDictr[pp]=i

oList = sorted(sDict.values())
nOfElements = len(oList)

sol = ''
count = 0
for i in oList:
    sol+=str(sDictr[i])
    count += 1
    if count != nOfElements:
        sol+=str(", ")
print(sol)
```

## Task 6

```
class trie:
    def __init__(self,charno=27):
        self.charno=charno
        self.root=0;
        self.tree=[[-1]*charno]
        self.end=[-1]               
        self.no_of_patters=0
        self.node=1

    def insert(self,pat):
        self.no_of_patters+=1
        cur=self.root
        for i,x in enumerate(pat):
            ind = ord(x) - ord('A')
            if self.tree[cur][ind]==-1:
                self.tree[cur][ind]=self.node
                self.tree.append([-1]*self.charno)
                cur=self.node
                self.node+=1
                self.end+=[-1]
            else:
                cur=self.tree[cur][ind]
        self.end[cur] = self.no_of_patters

    def search(self,pat):
        ret = []
        cur=self.root
        for i,x in enumerate(pat):
            ind = ord(x) - ord('A')
            if self.tree[cur][ind]==-1:
                break
            cur = self.tree[cur][ind]
            if self.end[cur]!=-1:
                ret.append(self.end[cur])

        return ret

    def printTrie(self, cur=-1, par=-1, sym=-1):
        if cur==-1:
            cur=self.root
        for i in range(self.charno):
            if self.tree[cur][i]!=-1:
                self.printTrie(self.tree[cur][i],cur,i)
        if par!=-1:
            print(str(par)+'->'+str(cur)+':',chr(sym+ord('A')))
```

```
patterns = []
for i in range(3):
    dd = input()
    patterns.append(dd)

tr_obj = trie()
for pat in patterns:
  tr_obj.insert(pat)
tr_obj.printTrie()
```
