## Task 1

```
def BWT(text):
    bwt = []
    for i,_ in enumerate(text):
        bwt.append(text[i:]+text[:i])
    bwt = sorted(bwt)
    answer = ''
    for v in bwt:
        answer+=(v[-1])
    return answer
```

```
text = input()
sol = BWT(text)
print(sol)
```

## Task 2

```
def uniqueToSeparate(text):
    a,c,t,g=0,0,0,0
    for j,i in enumerate(text):
        if i=='A':
            a+=1
            text[j]='A'+str(a)
        elif i=='G':
            g+=1
            text[j]='G'+str(g)
        elif i=='C':
            c+=1
            text[j]='C'+str(c)
        elif i=='T':
            t+=1
            text[j]='T'+str(t)
    return text
```

```
def InverseBWT(text):
    list_text = list(text)
    list_text = uniqueToSeparate(list_text)

    stt = sorted(text)
    
    stt = uniqueToSeparate(stt)

    answer = []
    answer.append('$')
    for i in range(len(stt)-1):
        idx = stt.index(answer[-1])
        answer.append(list_text[idx])
    return answer
```

```
text = input()

in_ans = InverseBWT(text)
sol=''
for ch in in_ans[::-1]:
    sol+=str(ch[0])
print(sol)
```

## Task 3

```
def l2f(transform, ii):
    list_text = list(transform)
    list_text = uniqueToSeparate(list_text)

    stt = sorted(transform)
    
    stt = uniqueToSeparate(stt)

    return stt.index(list_text[ii])
```

```
transform = input()
i = int(input())
answer = l2f(transform, i)
print(answer)
```
